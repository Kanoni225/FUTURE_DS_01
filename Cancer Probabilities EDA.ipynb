```python
#Probability of cancer
sns.countplot(data=Cancer, x='Probability of Cancer')
plt.show()
#Mode probability of cancer is 0.4
```

```python
#Bivariate Analysis
#The relationship between Smoking Habit and Drinking Habit
pd.crosstab(Cancer['Smoking Habit'], Cancer['Drinking Habit'])
#Smoking and drinking habits are directly proportional. The more the heavy smokers the more the frequent drinkers.
```

```python
#Relationship between Smoking Habit and Probabitlity of Cancer
sns.barplot(data=Cancer, x='Smoking Habit', y='Probability of Cancer')
plt.title('Relationship between Smoking Habit and the Probability of Cancer')
plt.show()
#The heavy smokers have the highest probability of cancer while the non-smokers have the least probability of cancer.
```

```python
#Relationship between Drinking habit and Probability of Cancer
sns.barplot(data=Cancer, x='Drinking Habit', y='Probability of Cancer')
plt.title('Relationship between Drinking Habit and Probability of Cancer')
plt.show()
#The frequent drinkers have the highest probability of cancer while those who do not drink have the lowest probability of cancer.
```

```python
#Relationship between Biking Habit and Probability of Cancer
sns.barplot(data=Cancer, x='Biking Habit', y='Probability of Cancer')
plt.title('Relationship between Biking Habit and Probability of Cancer')
plt.show()
#Those who have a low biking habit have the highest probability of cancer while those who have a high biking habit have the lowest probability of Cancer
```

```python
#Relationship between Walking habit and Probability of Cancer
sns.barplot(data=Cancer, x='Walking Habit', y='Probability of Cancer')
plt.title('Relationship between Walking Habit and Probability of Cancer')
plt.show()
#Those with the highest probability of cancer are those with the lowest walking habits while those with the lowest probability of cancer have the highest walking habits.
```

```python
#Relationship between Jogging Habit and Probability of Cancer
sns.barplot(data=Cancer, x='Jogging Habit', y='Probability of Cancer')
plt.title('Relationship between Jogging Habit and Probability of Cancer')
plt.show()
#The lower the jogging habit the higher the probability of cancer and the higher the jogging habit the lower the probability of cancer.
```
